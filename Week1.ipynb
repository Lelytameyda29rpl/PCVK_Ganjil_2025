# Langkah (Step) 1:


*   Numpy adalah pustaka manipulasi array, yang digunakan untuk aljabar linear, transformasi Fourier, dan kemampuan angka acak.

*   Matplotlib adalah pustaka yang menghasilkan figur dan menyediakan perangkat antarmuka pengguna grafis.







```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # untuk menampilkan gambar
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

# Langkah (Step) 2:

Pada langkah ini kita akan memebaca image dari URLs, menampilkannya menggunakan OpenCV, dengan memeperhatikan ada perbedaan ketika membuka image dalam RGB dan BGR. Channel Warna default untuk OpenCV adalah BGR.

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placecats.com/800/571"]


# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
    image = io.imread(url)                           # membaca gambar
    image = cv.resize(image, (0,0), fx=0.5, fy=0.5)  # mengubah ukuran gambar menjadi setengah ukuran
    image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # mengubah warna menjadi RGB
    final_frame = cv.hconcat((image, image_2))       # gambar gabungan
    cv2_imshow(final_frame)                          # menampilkan gambar
    print('\n')
```

# Langkah (Step) 3:

Melihat ukuran file image.

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

# Langkah (Step) 4:

Mengakses pixel dengan memeberi garis diagonal menyilang.

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horisontal ditengah image
for y in range (lebar):
    image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
